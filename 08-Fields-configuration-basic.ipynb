# Field configs (Aliases)

```python
from pydantic import BaseModel, ValidationError, Field
```

usually in frontend or json datas names are camelCase and you can handel it here

```python
class Data(BaseModel):
    first_name: str = Field(alias='firstName')
    last_name: str = Field(alias='lastName')
    age: int = Field(alias='Age', default=18)
```

```python
data = {
    "firstName": 'parshan',
    'lastName': 'Smith',
    'Age': 10
}

m = Data.model_validate(data)
m
```

```python
data = {
    "firstName": 'parshan',
    'lastName': 'Smith',
}

m = Data.model_validate(data)
m
```

but alias is'n going to work in while serializition

```python
m.model_dump()
```

You saw it was our names that we defined not alias but you can get alias in dump too like this:

```python
m.model_dump(by_alias=True)
```
