# Custom Serializing


````python
from pydantic import BaseModel, field_serializer, FieldSerializationInfo
from datetime import datetime
````

here we want to see how to customize your outputs by field serializer.


````python
class Event(BaseModel):
    name: str
    time: datetime

    @field_serializer("time", when_used="always")
    def parse(self, value) -> str:
        return f"Event {self.name} is created at {value.isoformat()}"
````

Note: there is so many other when_used parameter too but most usefuls are:

```always```: for any model dump.

```json```: for any json model dump.

```unless-none```: for any model dump but it just works when value be not none

```json-unless-none```: for json model dump and also value must not be none

````python
m = Event(name='worldcup', time='2026-01-01T12:00:00')
m
````

We don't see the customize that we applied because it just happen with model dump.

````python
m.model_dump()
````

and another example:

Can you make card nubers liuke this:
```5022-****-****-9064```?

but pydantic can so let's make it.!!

````python
class Payment(BaseModel):
    amount: int
    card_number: str
    @field_serializer('card_number')
    def card_srializer(self, value:str, info:FieldSerializationInfo):
        context = info.context.get('role') # check README.md for more information
        if context == 'user':
            return f'{value[:4]}-****-****-{value[-4:]}'
        return value
````

Now test:

````python
m = Payment(amount=1_000_000_000, card_number='5022123473339064')
m.model_dump(context={'role':'user'})
````

and Admin:

````python
m.model_dump(context={'role':'admin'})
````
