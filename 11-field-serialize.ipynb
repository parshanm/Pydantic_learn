# Custom Serializing


````python
from pydantic import BaseModel, field_serializer
from datetime import datetime
````

here we want to see how to customize your outputs by field serializer.


````python
class Event(BaseModel):
    name: str
    time: datetime

    @field_serializer("time", when_used="always")
    def parse(self, value) -> str:
        return f"Event {self.name} is created at {value.isoformat()}"
````

Note: there is so many other when_used parameter too but most usefuls are:

```always```: for any model dump.

```json```: for any json model dump.

```unless-none```: for any model dump but it just works when value be not none

```json-unless-none```: for json model dump and also value must not be none

````python
m = Event(name='worldcup', time='2026-01-01T12:00:00')
m
````

We don't see the customize that we applied because it just happen with model dump.

````python
m.model_dump()
````
