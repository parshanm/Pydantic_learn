# Advanced alias configs

```python
from pydantic import BaseModel, Field, ConfigDict, AliasChoices
```

## serialization aliases

if you want to just change out going alias you can config serialization alias

```python
class Model(BaseModel):
    model_config = ConfigDict(populate_by_name=True)

    first_name: str = Field(alias='firstName', serialization_alias='givenName')
```

```python
m = Model(firstName='pasrshan')
m
```

Now watch model dump

```python
m.model_dump(by_alias=True)
```

## Validation alias

It's really looks like the previous one but it's for model validation

```python
class Data(BaseModel):
    model_config = ConfigDict(populate_by_name=True)

    first_name: str = Field(validation_alias='FirstName', serialization_alias='givenName')
```

```python
x = Data(FirstName='parshan')
x
```

the validation alias just works in validation step✅

```python
x.model_dump(by_alias=True)
```

### Alias choices

alias choices give you the option to set two aliases for a field (Thast's really fantastic)

```python
class Person(BaseModel):
    model_config = ConfigDict(populate_by_name=True)

    first_name: str = Field(
        validation_alias=AliasChoices('firstName', 'givenName'),
        serialization_alias='givenName'
    )
```

The first alias that model looks for is the first one that given

```python
data = {
    'firstName':'parshan'
}

data_2 ={
    'firstName':'Isaac'
}
```

```python
m_1  = Person.model_validate(data)
m_1.model_dump()
```

```python
m_2 = Person.model_validate(data_2)
m_2.model_dump()
```
