## 01-strict

for this session you need to know some basics about python check them if you know them you can pass them

````python
from typing import Annotated
from uuid import UUID
````

with ```annotated``` we can define types to check our conditions for input data.

And we use ```UUID``` to generate somerandom unique IDs.

now we want generate our types with ```Annotated``` to make our code less & better

````python
from pydantic import BaseModel, Strict

class Model(BaseModel):
    id: int
````

we saw in lax mode if we send a number to model in str pydantic were turned that to a int.

````python
c = Model(id='12')
c.model_dump()
````

and we saw we can set our model strict or just a field but here we do it with less code

````python
class Model(BaseModel):

    id: Annotated[int, Strict()] # works like: Annotated[type, conditions]
````

if we test now out field must be strict

````python
c = Model(id='12')
c.model_dump()
````

## 02-AllowInfNan

this can help you to handle math methods in ypur API

````python
from pydantic import AllowInfNan

class Model(BaseModel):
    value: Annotated[float, AllowInfNan(True)]
````

````python
from math import nan

c = Model(value=nan)
c.model_dump()
````

just like this!!

## 03-string-constraintrs

````python
from pydantic import StringConstraints

class Model(BaseModel):
    name: Annotated[str, StringConstraints(strip_whitespace=True, min_length=2, max_length=10)] # or any config you need
````

````python
m = Model(name='    parshan    ')
m.model_dump()
````

this is one of the funniest methods in pydantic

## 04-import-string

with this method you can import a function or anything and save it in a varriable

````python
from pydantic import ImportString

class Model(BaseModel):
    func: ImportString
````

````python
m = Model(func='math.sin')

print(f'this is the real type: {type(m.func)}')
print(f'this is the usage: {m.func(0)}')
````

## 05-UUID-version

````python
from pydantic.types import UuidVersion

class Model(BaseModel):
    id: Annotated[UUID, UuidVersion(4)] # this just accept uuid version 4
````

## 06-JSON

````python
from pydantic.types import Json

class Model(BaseModel):
    value: Json[list[int]]
````

you can convert any json data to real python standard types

````python
m = Model(value='[1,3,4,5,6,7,8,9]')
m.model_dump()
````
