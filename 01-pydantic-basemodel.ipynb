```python
from pydantic import BaseModel, ValidationError
```

# define a class with pydantic

```python
class User(BaseModel):
    username: str
    password: str
    age: int
    @property
    def fullname(self):
        return f"{self.username} is {self.age} years old"
```

# defining some user objects

```python
parshan = User(username="parshan", password="1234", age=15)
IR = User(username="Islamic Republic", password="die", age=47)
```

```python
parshan.fullname
```

# Deserialization

```python
data ={
    "username": "Donald J. Trump",
    "password": "MAGA",
    "age": 79
}
Donald = User.model_validate(data)
Donald.fullname
```

```python
data = '''
{
    "username": "Benjamin Netanyahu",
    "password": "lion",
    "age": 76
}
'''
bibi = User.model_validate_json(data)
bibi.fullname
```

# Serialization

```python
parshan.model_dump()
```

```python
IR.model_dump_json()
```
