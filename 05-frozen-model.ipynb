```python
from pydantic import BaseModel, ValidationError, ConfigDict
```

# Mutable and Immutable models

```python
class Model(BaseModel):
    name: str
    age: int
    height: float
    weight: float
```

I can change my data easily like that we saw before, but I'm going to make my model immutable

```python
class Model(BaseModel):
    model_config = ConfigDict(frozen=True)

    name: str
    age: int
    height: float
    weight: float
```

Now I cannot change my model anyway

```python
m = Model(name="model", age=20, height=100, weight=100)
m
```

```python
m.age = 10
```

# Making cache of model (for frozen only)

```python
d = {
    'data_1':m,
    'data_2':Model(name="model_1", age=20, height=100, weight=100),
}
```

we call it cache that stores in RAM😊
