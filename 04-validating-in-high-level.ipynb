```python
from pydantic import BaseModel, ValidationError, ConfigDict
```

# Validating default values

```python
class User(BaseModel):
    model_config = ConfigDict(validate_default=True)

    username: str = None
    password: str = None
    email: str = None
```

```python
try:
    User()
except ValidationError as e:
    print(e)
```

Yeah it makes an error 🥸😊

# Data Validation for assigment data

Now if i make an usual model:

```python
class User(BaseModel):
    username: str
    password: str
    email: str
```

```python
user = User(username="admin", password="", email="")
user
```

But if I assign data like this:

```python
user.email = 100
```

Yep! It's not going to raise an exception,
But we can fix this with ConfigDict

```python
class User(BaseModel):
    model_config = ConfigDict(validate_assignment=True)

    username: str
    password: str
    email: str
```

```python
user = User(username="admin", password="", email="")
user
```

```python
try:
    user.email = 100
except ValidationError as e:
    print(e)
```
