```python
from pydantic import BaseModel, ValidationError, ConfigDict
```

# Ignoring extra Fields

```python
class Data(BaseModel):
    fields_1: int = 0
```

```python
m = Data(fields_1=5, extra=10)
m
```

Nope extra does not exist because pydantic ignore extra fields by default

Just like this:

```python
class Data(BaseModel):
    model_config = ConfigDict(extra='ignore')

    fields_1: int = 0
```

```python
m = Data(fields_1=5, extra=10)
m
```

See! the output uis exactly the same

But it's changeable at first raising validation error with forbid

```python
class Data(BaseModel):
    model_config = ConfigDict(extra='forbid')

    fields_1: int = 0
```

```python
try:
    m = Data(fields_1=5, extra=10)
    m
except Exception as ex:
    print(ex)
```

But we also can get and store them

```python
class Data(BaseModel):
    model_config = ConfigDict(extra='allow')

    fields_1: int = 0
```

```python
m = Data(fields_1=5, extra=10)
m
```
