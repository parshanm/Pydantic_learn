# Alias generators

```python
from pydantic import BaseModel, ValidationError, Field, ConfigDict
from pydantic.alias_generators import to_camel, to_pascal, to_snake
```

## Using pre-built alias generators

```python
class Model(BaseModel):
    model_config = ConfigDict(alias_generator=to_camel)

    id_: int = Field(alias='id') # override aliases
    first_name: str
    last_name: str
```

```python
m = Model(id=1, firstName='aziz', lastName='hoooray!!')
m
```

```python
try:
    Model(id=1, first_name='parshan', last_name='Smith')
except ValidationError as ex:
    print(ex)
```

Yeah here you can't use first_name, but there is a way to use them like this:

```python
class Model(BaseModel):
    model_config = ConfigDict(alias_generator=to_camel, populate_by_name=True)

    id_: int = Field(alias='id')
    first_name: str
    last_name: str
```

```python
try:
    Model(id=1, first_name='parshan', last_name='Smith')
    Model(id=1, firstName='jack', lastName='Smith')
except ValidationError as ex:
    print(ex)
```

and it worked without any error✅

## Making your own alias generator

```python
def to_upper(in_str: str)-> str:
    return in_str.upper()
```

```python
class Model(BaseModel):
    model_config = ConfigDict(alias_generator=to_upper)

    id_: int 
    first_name: str
    last_name: str
```

```python
m = Model(ID_=1, FIRST_NAME='lisa', LAST_NAME='su')
m
```
