# What is Enum?

```python
from enum import Enum
```

```python
class Color(str, Enum):
    blue = 'Blue'
    red = 'Red'
    orange = 'Orange'
    green = 'Green'
    yellow = 'Yellow'
    purple = 'Purple'
    black = 'Black'
    white = 'White'
    gray = 'Gray'
```

```python
Color.red
```

Enums have value so we can do this easily there is no key for them (name instead)

```python
Color.orange.value
```

```python
Color.orange.name
```

```python
type(Color.red), type(Color.red.value)
```

# Link Enums and Pydantic

```python
from pydantic import BaseModel, ConfigDict, ValidationError
```

```python
class Data(BaseModel):
    color: Color
```

## Let's test

just send the value of Enum ad pydantic covert it (Fantastic)

```python
data = """
{
    "color": "Red"
}
"""
Data.model_validate_json(data)
```

let's make a wrong choice

```python
data = """
{
    "color": "Cyan"
}
"""
try:
    Data.model_validate_json(data)
except ValidationError as ex:
    print(ex)
```

Yes pydantic knows the correct values

But saving a python class in database is not a good way let's fix it

```python
class Data(BaseModel):
    model_config = ConfigDict(use_enum_values=True)

    color: Color
```

Done the problem has been solved

```python
data = """
{
    "color": "Red"
}
"""
Data.model_validate_json(data)
```
