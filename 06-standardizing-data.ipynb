```python
from pydantic import BaseModel, ValidationError, ConfigDict
```

# standardizing data

you can save numbers as str to make easy to save or show them:

```python
class Model(BaseModel):
    model_config = ConfigDict(coerce_numbers_to_str=True)
    field_1: str
```

```python
m = Model(field_1=10)
m
```

## Also you can delete white spaces in string too.

```python
class Model(BaseModel):
    model_config = ConfigDict(str_strip_whitespace=True)
    field_1: str
```

```python
m = Model(field_1='    pydantic\t')
m_1 = Model(field_1='pydantic\t      ')
m, m_1
```

## Or you can make str lowercase

```python
class Model(BaseModel):
    model_config = ConfigDict(str_to_lower=True)
    field_1: str
```

```python
m = Model(field_1='Pydantic')
m
```

## Or reverse!

```python
class Model(BaseModel):
    model_config = ConfigDict(str_to_upper=True)
    field_1: str
```

```python
m = Model(field_1='Pydantic')
m
```
